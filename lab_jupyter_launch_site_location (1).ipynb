```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Import required libraries
from js import fetch
import io
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Load dataset
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Keep relevant columns
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Create map centered at average location
mean_lat = spacex_df['Lat'].mean()
mean_long = spacex_df['Long'].mean()
success_map = folium.Map(location=[mean_lat, mean_long], zoom_start=4)

# Initialize MarkerCluster
marker_cluster = MarkerCluster().add_to(success_map)

# Add markers to the cluster
for index, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    site = row['Launch Site']
    launch_result = 'Success' if row['class'] == 1 else 'Failure'
    color = 'green' if row['class'] == 1 else 'red'
    
    folium.Marker(
        location=coord,
        popup=f"{site} | {launch_result}",
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)

# Display the clustered map
success_map
```

```python
from js import fetch
import io
import pandas as pd
from math import radians, cos, sin, sqrt, atan2

# Fetch and load SpaceX launch data with geocoordinates
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Prepare launch sites dataframe
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Define haversine distance function
def haversine_distance(coord1, coord2):
    R = 6371.0  # Earth radius in km
    lat1, lon1 = map(radians, coord1)
    lat2, lon2 = map(radians, coord2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Example: Calculate distances from CCAFS LC-40 to some proximities
launch_site = 'CCAFS LC-40'
launch_row = launch_sites_df[launch_sites_df['Launch Site'] == launch_site].iloc[0]
launch_coord = (launch_row['Lat'], launch_row['Long'])

coast_coord = (28.562, -80.567)
orlando_coord = (28.5383, -81.3792)
nasa_jsc_coord = (29.559684888503615, -95.0830971930759)

dist_to_coast = haversine_distance(launch_coord, coast_coord)
dist_to_city = haversine_distance(launch_coord, orlando_coord)
dist_to_jsc = haversine_distance(launch_coord, nasa_jsc_coord)

print(f"Distance from {launch_site} to Coast: {dist_to_coast:.2f} km")
print(f"Distance from {launch_site} to Orlando: {dist_to_city:.2f} km")
print(f"Distance from {launch_site} to NASA JSC: {dist_to_jsc:.2f} km")
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
from math import sin, cos, sqrt, atan2, radians
import folium
from folium.features import DivIcon

# Launch site coordinate
launch_site_coord = (28.562, -80.577)  # Example: CCAFS LC-40

# Proximity points with their names and coordinates
proximities = {
    "Closest Coastline": (28.56367, -80.57163),
    "Closest City (Orlando)": (28.5383, -81.3792),
    "Closest Railway": (28.560, -80.580),  # Example coord, replace with real data
    "Closest Highway": (28.565, -80.575),  # Example coord, replace with real data
}

# Calculate distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1)*cos(lat2)*sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Create map centered at launch site
site_map = folium.Map(location=launch_site_coord, zoom_start=13)

# Add launch site marker
folium.Marker(
    location=launch_site_coord,
    popup="Launch Site",
    icon=folium.Icon(color='green', icon='rocket')
).add_to(site_map)

# Add markers for proximities with distance labels and connecting lines
for name, coord in proximities.items():
    dist = calculate_distance(launch_site_coord[0], launch_site_coord[1], coord[0], coord[1])
    
    # Add marker with distance label
    folium.Marker(
        location=coord,
        popup=f"{name}: {dist:.2f} km",
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12pt; color: #d35400;"><b>{dist:.2f} KM</b></div>',
        )
    ).add_to(site_map)
    
    # Draw line from launch site to proximity point
    folium.PolyLine(
        locations=[launch_site_coord, coord],
        weight=2,
        color='blue',
        opacity=0.7
    ).add_to(site_map)

site_map
```
